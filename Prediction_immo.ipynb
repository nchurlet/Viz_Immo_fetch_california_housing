# Etude du prix de l'immobilier à Boston

Le but de cet exercice est d'utiliser le jeu de données sur l'immobilier à Boston comme dataset "fil rouge" afin de tester et acquérir des compétences.

Importation des librairies nécessaires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Affichage du dataset

```python
from sklearn import datasets
data = datasets.fetch_california_housing()
```

```python
dataset = pd.DataFrame(columns=data["feature_names"], data=data.data)  
dataset = pd.concat([dataset,pd.DataFrame({"Price":data.target})], axis=1)
dataset.head()
```

```python
def setCoordonees(setOfData):
    y = dataset.Latitude.values
    x = dataset.Longitude.values
```

```python
setCoordonees(dataset)
graph = sns.jointplot(x=x,y=y,height=50)
```

```python
z = dataset.Price.values
zMax = z.max()
zMax
```

```python
isDebug = False
preview = dataset.head(1000)
if(isDebug):
    datasetUsed=preview
else:
    datasetUsed=dataset
```

```python
setCoordonees(datasetUsed)
sns.relplot(x="Longitude", y='Latitude', hue="Price", height= 25, data=datasetUsed)
```
